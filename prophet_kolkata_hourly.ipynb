```python
origin = '2021-01-01 00:00:00'
start = '2024-01-01 00:00:00'
end = '2024-01-01 23:00:00'
time_period = pd.date_range(start,end,freq='h')
val['From Date'] = time_period
```

```python
tot = pd.DataFrame()
tot['PM2.5 (µg/m³)'] = pd.concat([df['PM2.5 (µg/m³)'],val['PM2.5']], ignore_index = True, axis = 0)
```

```python
tot['Timestamp'] = pd.concat([df['Timestamp'],val['From Date']], ignore_index = True, axis = 0)
```

```python
tot.set_index('Timestamp',drop=False,inplace=True)
```

```python
y_preds = []
ts = tot.loc[origin:end]['PM2.5 (µg/m³)'].asfreq("h")
ts
```

```python
for cur_hr in time_period.shift(-1):
    data = pd.DataFrame()
    data['ds'] = tot.loc[origin:cur_hr]['Timestamp'].asfreq('h')
    data['y'] = ts.loc[origin:cur_hr].asfreq('h')
    m = Prophet()
    m.fit(data)
    future = pd.DataFrame()
    future['ds'] = [cur_hr]
    forecast = m.predict(future)
    y_pred = forecast['yhat']
    print(y_pred.iloc[0])
    y_preds.append(y_pred.iloc[0])
y_preds = pd.Series(y_preds, index=time_period)
mae = (y_preds - ts[start:end]).abs().mean()
```

```python
mae
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(val['PM2.5'].values,label='original_values',marker='o')
ax.plot(y_preds.values,label='predicted_values',marker='o')
ax.legend()
plt.xlabel('Hours in Jan 1, 2024 from 00:00 to 23:00')
plt.ylabel('PM 2.5 Concentration in Kolkata')
```
