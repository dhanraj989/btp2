```python
# Plot the data
import plotly.express as px
import plotly.graph_objects as go
fig = px.line(df, x='Timestamp', y="PM2.5 (µg/m³)")
fig.show()
```

```python
# Make a copy of the original DataFrame
df_copy = df.copy()

# Mark the missing values before imputation
imputed_indices = df_copy[df_copy['PM2.5 (µg/m³)'].isnull()].index

# Apply the rolling mean imputation method
df_copy['PM2.5 (µg/m³)'] = df_copy['PM2.5 (µg/m³)'].fillna(df_copy['PM2.5 (µg/m³)'].rolling(window=8, min_periods=1).mean().shift(1))

fig = px.line(df_copy, x='Timestamp', y="PM2.5 (µg/m³)")

# Add points where data was imputed with red color
fig.add_trace(go.Scatter(x=df_copy.loc[imputed_indices, 'Timestamp'],
                         y=df_copy.loc[imputed_indices, 'PM2.5 (µg/m³)'],
                         mode='markers',
                         marker=dict(color='red'),
                         name='Imputed Data'))

# Set labels
fig.update_xaxes(title_text='Time')
fig.update_yaxes(title_text='PM2.5 (µg/m³)')

# Show the figure
fig.show()
```
