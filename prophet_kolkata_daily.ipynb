```python
origin = '2021-01-01'
start = '2024-01-01'
end = '2024-01-31'
time_period = pd.date_range(start,end,freq='D')
val['From Date'] = time_period
```

```python
tot = pd.DataFrame()
tot['PM2.5 (µg/m³)'] = pd.concat([df['PM2.5 (µg/m³)'],val['PM2.5']], ignore_index = True, axis = 0)
```

```python
tot['Timestamp'] = pd.concat([df['Timestamp'],val['From Date']], ignore_index = True, axis = 0)
```

```python
tot.set_index('Timestamp',drop=False,inplace=True)
```

```python
y_preds = []
ts = tot.loc[origin:end]['PM2.5 (µg/m³)'].asfreq("D")
ts
```

```python
for cur_hr in time_period.shift(-1):
    data = pd.DataFrame()
    data['ds'] = tot.loc[origin:cur_hr]['Timestamp'].asfreq('D')
    data['y'] = ts.loc[origin:cur_hr].asfreq('D')
    m = Prophet()
    m.fit(data)
    future = pd.DataFrame()
    future['ds'] = [cur_hr]
    forecast = m.predict(future)
    y_pred = forecast['yhat']
    print(y_pred.iloc[0])
    y_preds.append(y_pred.iloc[0])
y_preds = pd.Series(y_preds, index=time_period)
mae = (y_preds - ts[start:end]).abs().mean()
```
